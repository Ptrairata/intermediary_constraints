# Key Variable Definitions for Regression Analysis

This notebook documents and explains the construction and interpretation of key variables used in the regression analysis of intermediary constraints in financial markets.  
The variables are grouped into two categories:  

1. **Domestic Constraint Proxies** – Measures capturing balance sheet costs and liquidity conditions faced by U.S.-based intermediaries.  
2. **Foreign Constraint Proxies** – Measures capturing deviations from covered interest parity and funding stress in major foreign exchange markets.

For each variable, we provide:
- **Definition** – Market context and construction method.  
- **Economic Interpretation** – Why the variable serves as a proxy for intermediary constraints.  
- **Commentary** – Observations on historical patterns and notable stress episodes.  

The objective is to ensure transparency in variable selection, provide intuition for their use in regression analysis, and highlight linkages between domestic and foreign funding markets.

## Domestic Constraint Proxies

### GCF–Repo Survey Spreads

This variable measures the premium that dealers are willing to pay in the General Collateral Finance (GCF) repo market relative to surveyed repo rates.  

The GCF repo market — cleared through the Fixed Income Clearing Corporation (FICC) — allows dealers to net their repo and reverse repo positions against each other.  
Because these offsetting trades are centrally cleared, collateral movements are reduced or eliminated, lowering balance sheet usage.  

A positive spread of GCF rates over comparable surveyed repo rates reflects the value of this *balance sheet netting benefit*.  
A higher spread indicates greater willingness to pay for balance sheet relief, suggesting tighter constraints on dealer intermediation capacity.


```python
df1 = pd.read_csv("../data/GCF_2009_present.csv", parse_dates = ['Date'])
df2 = pd.read_csv("../data/repo_survey_2014_present_clean.csv", parse_dates = ['Date'])

gcf_survey = pd.merge(df1,df2, on='Date', how = 'inner')
gcf_survey = gcf_survey.rename(columns={'Rate':'Survey'})
gcf_survey['diff']= (gcf_survey['GCF']-gcf_survey['Survey'])*100 #bps
gcf_survey.to_csv("../data/GCF_survey.csv", index = False)
```

```python
# Plot
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(gcf_survey['Date'], gcf_survey['diff'], label='GCF - Survey Difference', color='blue')



# Formatting
ax.axhline(0, color='black', linewidth=1, linestyle='--')
ax.set_xlabel('Date')
ax.set_ylabel('Difference (bps)')
ax.set_title('Difference between GCF and Survey Rates Over Time')

# Format y-axis as percentage
## ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))  # 1.0 = already in decimal form

ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

#### Commentary

The spread typically remains close to zero, with occasional upward spikes.  
Notable increases occurred in early 2019 and again in March 2020 at the onset of the COVID-19 pandemic, reflecting periods of heightened balance sheet constraints in the repo market.


### IORB–SOFR Spread

The spread between the Interest on Reserve Balances (IORB) and the Secured Overnight Financing Rate (SOFR) represents a textbook arbitrage in a frictionless market.  

In principle, intermediaries could borrow cash in the repo market at SOFR and deposit it at the Federal Reserve to earn IORB, capturing the spread as risk-free profit.  

In practice, exploiting this opportunity requires expanding their balance sheet—taking on additional assets and liabilities—which is costly due to regulatory leverage ratios, capital requirements, and internal risk limits.  

A larger IORB–SOFR spread therefore reflects a higher implicit cost of balance sheet expansion for intermediaries.


```python
df1 = pd.read_csv("../data/SOFR_2014_present.csv", parse_dates = ['Date'])
df2 = pd.read_csv("../data/IORB_2021_present.csv", parse_dates = ['Date'])

IORB_SOFR = pd.merge(df1,df2, on='Date', how = 'inner')

IORB_SOFR['diff']= (IORB_SOFR['IORB']-IORB_SOFR['SOFR'])*100
IORB_SOFR.to_csv("../data/IORB_SOFR.csv", index = False)
```

```python
# Plot
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(IORB_SOFR['Date'], IORB_SOFR['diff'], label='IORB - SOFR', color='red')



# Formatting
ax.axhline(0, color='black', linewidth=1, linestyle='--')
ax.set_xlabel('Date')
ax.set_ylabel('Difference (bps)')
ax.set_title('Difference between IORB and SOFR Over Time')

# Format y-axis as percentage
# ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))  # 1.0 = already in decimal form

ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

#### Commentary

Because the IORB rate is adjusted infrequently, the IORB–SOFR spread tends to move in relatively stable increments.  
However, from mid-2024 onward, the spread has exhibited noticeably higher volatility, suggesting changing market conditions or shifts in liquidity dynamics.  

Next, we examine the correlation between this measure and the GCF–Repo Survey Spread, as both are intended to capture aspects of domestic intermediary constraints.


```python
df = pd.merge(gcf_survey,IORB_SOFR, on= 'Date', how='inner')
df = df.rename(columns = {'diff_x':'gcf_survey','diff_y':'iorb_sofr'})
df.head()
```

```python
r_squared = df['gcf_survey'].corr(df['iorb_sofr']) ** 2
print(r_squared)
```

### Foreign Constraint Proxies

#### Cross-Currency Basis

The cross-currency basis measures the deviation from Covered Interest Parity (CIP) for a given maturity.  
A positive basis (from the perspective of foreign currency holders) indicates that borrowing U.S. dollars via the FX swap market is cheaper than implied by CIP, while a negative basis implies it is more expensive.  

In a frictionless market, such deviations would present arbitrage opportunities—similar in spirit to the IORB–SOFR spread—where intermediaries could lock in risk-free profits by trading in FX swaps and cash markets.  

Persistent non-zero basis values reflect limits to arbitrage, often due to balance sheet constraints, regulatory capital requirements, or funding market frictions that deter intermediaries from fully exploiting these opportunities.  

This analysis focuses on the three most-traded currency pairs against the U.S. dollar: GBP/USD, EUR/USD, and USD/JPY.


```python
# 3-Month JPY/USD Basis
df_jpybs = pd.read_csv("../data/JYBS2021_present.csv", parse_dates = ["Date"])
df_jpybs.head()
```

```python
# Plot
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df_jpybs['Date'], df_jpybs['JYBS3M'], label='Basis USD/JPY 3M (bps)', color='blue')



# Formatting
ax.axhline(0, color='black', linewidth=1, linestyle='--')
ax.set_xlabel('Date')
ax.set_ylabel('Basis (bps)')
ax.set_title('Cross currency Basis USD/JPY 3M')


ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_gbpbs = pd.read_csv("../data/GBPUSD_BS_2021_present.csv", parse_dates = ['Date'])
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df_gbpbs ['Date'], df_gbpbs ['BPBS_3MO'], label='Basis GBP/USD 3M (bps)', color='blue')



# Formatting
ax.axhline(0, color='black', linewidth=1, linestyle='--')
ax.set_xlabel('Date')
ax.set_ylabel('Basis (bps)')
ax.set_title('Cross currency Basis GBP/USD 3M')


ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_eubs = pd.read_csv("../data/EURUSD_BS_2021_present.csv", parse_dates = ['Date'])
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df_eubs ['Date'], df_eubs ['EUBS_3MO'], label='Basis EUR/USD 3M (bps)', color='blue')



# Formatting
ax.axhline(0, color='black', linewidth=1, linestyle='--')
ax.set_xlabel('Date')
ax.set_ylabel('Basis (bps)')
ax.set_title('Cross currency Basis EUR/USD 3M')


ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

#### Commentary

All three bases display broadly similar patterns over time:  
- The GBP and EUR bases remain persistently positive.  
- The JPY basis remains persistently negative.  

Spikes and troughs tend to occur simultaneously across currencies and align with movements in the domestic constraint proxies.  
This co-movement is not surprising—given the high degree of financial market integration, stress in one segment can propagate rapidly across funding markets through multiple channels.  

Next, we examine how the variations in each currency’s basis relate to one another.


```python
dfs = [df_jpybs, df_eubs, df_gbpbs]
df_bs = reduce(lambda left, right: pd.merge(left, right, on='Date', how='inner'), dfs)
df_bs[['JYBS3M','EUBS_3MO','BPBS_3MO']].corr()
```

We can see that GBP and EUR have very correlated bases, this is likely becasue those two currencies share much of the same intermediaries. JPY has less related for the reason: there is less interesection in term of intermediaries.
